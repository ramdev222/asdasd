```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
million=1000000
billion=1000000000
economy_data = pd.read_csv('Economy_Data.csv'
                           ,delimiter=';')
economy_data=economy_data.fillna({'Value':0.0,'CountryName':'Unknown'})

economy_data['Value in Million']=economy_data['Value']/million
economy_data['Value in Billion']=economy_data['Value']/billion
economy_data.head(10)
```

```python
plt.figure(figsize=(12,8))
vis=sns.relplot(x="Year", y="Value in Billion",  hue='CountryName',kind="line", data=economy_data)
vis.ax.set_title(label='Brics Ecomony Growth 1970-2020',fontsize=15)
plt.show();
```

```python
countries=economy_data.CountryName.unique()
for country in  countries:   
    if country != 'Unknown' :        
        economy_data_by_country=economy_data[economy_data['CountryName']==country]
        vis=sns.relplot(x="Year", y="Value in Billion", kind="line", data=economy_data_by_country)
        title_g=country+" Ecomony Growth 1970-2020"
        vis.ax.set_title(label=title_g,fontsize=15)
        plt.show();
```
